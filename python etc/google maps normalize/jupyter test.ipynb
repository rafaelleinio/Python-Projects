```python
df1 = pandas.read_excel("parse_addres.xlsx")
for i, row in df1.iterrows():
    address = row["string"]
    location = g.geocode(address,timeout=10)
    df1["state"][i] = location.raw['address_components'][4]['short_name']
    df1["city"][i] = location.raw['address_components'][3]['short_name']
    df1["address"][i] = location.raw['address_components'][1]['short_name'] + ", " + location.raw['address_components'][0]['short_name'] 
    df1["neighborhood"][i] = location.raw['address_components'][2]['short_name']
    df1["zipcode"][i] = location.raw['address_components'][6]['short_name']
    df1["lat"][i] = location.latitude
    df1["lng"][i] = location.longitude
    df1["enabled"][i]= "TRUE"
    df1["name"][i] = "Polo " + df1["neighborhood"][i] + " " +df1["city"][i] + " - " + df1["state"][i]
    if i == 10:
        break
df1
```
